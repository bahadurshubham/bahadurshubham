```python
#loading the datasets from kaggle on the dataframe
import pandas as pd
import numpy as np

housing = pd.read_csv('housing.csv')
housing['ocean_proximity'].value_counts()
```

```python
#randomizing the datasets
housing_df = housing.sample(n=len(housing) , random_state=1)
housing_df
```

```python
#making dummies of the original dataset
pd.get_dummies(housing_df['ocean_proximity']).head()
```

```python
#concatenating all the dummies and the and removing the ocean_proximity row
housing_final = pd.concat([housing_df.drop('ocean_proximity' , axis=1),
                          pd.get_dummies(housing_df['ocean_proximity'])]  , axis=1)
len(housing_final)
```

```python
#changing the output label from between to the last
housing_final = housing_final[['longitude','latitude',
                              'housing_median_age','total_rooms',
                              'total_bedrooms',	'population',
                              'households',	'median_income',
                              '<1H OCEAN','INLAND',
                              'ISLAND',	'NEAR BAY',
                              'NEAR OCEAN','median_house_value']]
```

```python
#dropping all the zaero rows and columns
housing_final = housing_final.dropna()
housing_final
```

```python
#splitting the data into training, validation and test data
train_df, val_df, test_df, = housing_final[:18000] , housing_final[18000:19318] , housing_final[19319:20433]
train_df.shape , val_df.shape , test_df.shape
```

```python
#splitting the training, validation and test datasets into inputs and outputs
X_train , Y_train = train_df.to_numpy()[: , :-1] , train_df.to_numpy()[: , -1]
X_val , Y_val = val_df.to_numpy()[: , :-1] , val_df.to_numpy()[: , -1]
X_test , Y_test = test_df.to_numpy()[: , :-1] , test_df.to_numpy()[: , -1] 

X_train.shape , Y_train.shape, X_val.shape , Y_val.shape, X_test.shape , Y_test.shape
```

```python
#now scaling the data and creating a function for the same 
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler().fit(X_train[: , :8])

def preprocessor(X):
  A = np.copy(X)
  A[: , :8] = scaler.transform(A[: , :8])
  return A

X_train ,X_val ,X_test = preprocessor(X_train), preprocessor(X_val), preprocessor(X_test)
#converting the scaled model into a dataframe
pd.DataFrame(X_train).head()
```

```python
#calculating the loss function from linearRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse

lr = LinearRegression().fit(X_train, Y_train)
mse(lr.predict(X_train) , Y_train , squared=False) , mse(lr.predict(X_val) , Y_val , squared=False)
```

```python
#implementing the KNeighbors regressor and then the loss function
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=10).fit(X_train , Y_train)
mse(knn.predict(X_train) , Y_train , squared=False) , mse(knn.predict(X_val) , Y_val, squared=False)
```

```python
#implementing the randomforestregressor and then the loss function
from sklearn.ensemble import RandomForestRegressor
rnn = RandomForestRegressor(max_depth=10).fit(X_train , Y_train)
mse(rnn.predict(X_train), Y_train , squared=False) , mse(rnn.predict(X_val) , Y_val , squared=False)
```

```python
#implementing the gradiantboostregressor and then the loss function
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor(n_estimators=170).fit(X_train , Y_train)
mse(gbr.predict(X_train) , Y_train , squared=False) , mse(gbr.predict(X_val) , Y_val , squared=False)
```

```python
#implementing the tensorflow framework to calculate the loss function
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

sample_nn = Sequential()
sample_nn.add(InputLayer((13,)))
sample_nn.add(Dense(256, 'relu'))
sample_nn.add(Dense(128, 'relu'))
sample_nn.add(Dense(64, 'relu'))
sample_nn.add(Dense(32, 'relu'))
sample_nn.add(Dense(16, 'relu'))
sample_nn.add(Dense(1, 'linear'))

opt = Adam(learning_rate=.3)
cp = ModelCheckpoint('models/sample_nn' , save_best_only=True)
sample_nn.compile(optimizer=opt, loss='mse' , metrics=[RootMeanSquaredError()])
sample_nn.fit(x=X_train , y=Y_train , validation_data=(X_val, Y_val) ,callbacks=[cp], epochs=10)
```
